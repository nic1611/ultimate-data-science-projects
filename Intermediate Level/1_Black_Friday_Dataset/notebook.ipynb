```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
df = pd.read_csv('train.csv')
df
```

```python
fig, ax =plt.subplots(ncols=2, nrows=3, figsize=(14,9))
ax = ax.flatten()
index = 0

for key in df[['Gender', 'Age', 'Occupation', 'City_Category', 'Stay_In_Current_City_Years', 'Marital_Status']].keys():
  sbn.barplot(data=df, x=key, y='Purchase', ax=ax[index])
  index += 1

plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=5.0)
```

```python
fig, ax =plt.subplots(ncols=3, nrows=2, figsize=(20,15))
ax = ax.flatten()
index = 0

for key in df[['Occupation', 'Product_Category_1', 'Product_Category_2', 'Product_Category_3', 'Purchase']].keys():
  sbn.boxplot(y=key, data=df, ax=ax[index])
  index += 1

plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=5.0)
```
